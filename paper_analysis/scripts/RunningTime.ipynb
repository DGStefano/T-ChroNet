```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
#### TIME ANALYSIS
time_results = pd.read_csv("/mnt/nas-safu/analysis/PhDsdigiove/method_coAcces/running_time.tsv" , sep = "\t")
time_results['time_min'] = time_results['estimated_time'] / 60
time_results['time_min'] = time_results['time_min'].astype(int)
```

```python
#sns.set_context("notebook", font_scale=1.2)
plt.figure(figsize=(9, 5))
sns.lineplot(data = time_results , x = 'row_name' , y = 'time_min' , hue = 'n_processors' , palette = sns.color_palette("rocket" , 8))
sns.scatterplot(data = time_results , x = 'row_name' , y = 'time_min' , hue = 'n_processors' , palette = sns.color_palette("rocket" , 8), legend=False)
plt.legend(title="Number of threads" , bbox_to_anchor=(1.02, 1), loc='upper left', borderaxespad=0 , frameon=False ,  fontsize=8, title_fontsize=10)
plt.ylabel("Time (min)" , fontsize=10)
plt.xlabel("Number of nodes", fontsize=10)
plt.xticks(fontsize=8)
plt.yticks(fontsize=8)
sns.despine()
# plt.savefig("/mnt/nas-safu/analysis/PhDsdigiove/method_coAcces/Threads_runningtime.png", dpi=300, bbox_inches='tight')
```
